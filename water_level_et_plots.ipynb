# Groundwater Level and ET plots 

The following plots are produced to visualize a monitoring well's groundwater level and ET over time. It is meant to help the viewer look for patterns between the two.

### Load in Data

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

### Import Data

```python
# Create a groundwater file path
wilcox_fp = os.path.join("/capstone/aridgw/data/merged_timeseries_1km/et_gw_merged_arizona_willcox_basin.csv")

# Read in groundwater data
wilcox_df = pd.read_csv(wilcox_fp)
wilcox_df.head()
```

```python
plt.figure(figsize=(12, 6))

# Plot the line as before
plt.plot(wilcox_df['YEAR'], wilcox_df['WaterLevel_m'], 'o-', 
         color='steelblue', linewidth=2, markersize=6, label='Water Level')

# Add area fill between the line and a reference level
reference_level = wilcox_df['WaterLevel_m'].min()  # Fill to the shallowest (best) level

plt.fill_between(wilcox_df['YEAR'], wilcox_df['WaterLevel_m'], reference_level, 
                 alpha=0.3, color='lightcoral', label='Groundwater Depletion')

plt.xlabel('Year', fontsize=18)
plt.ylabel('Depth to Groundwater (m)', fontsize=18)
plt.title('Depth to Groundwater Over Time In Wilcox, Arizona', fontsize=20)
plt.gca().invert_yaxis()  # Flip y axis
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=17)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(wilcox_df['YEAR'], wilcox_df['mean_ET_mm_yr'], 's-', label='Mean ET (mm/yr)', color='orange')
plt.xlabel('Year', fontsize=18)
plt.ylabel('Mean ET (mm/yr)', fontsize=18)
plt.title('Mean Evapotranspiration Over Time In Wilcox, Arizona', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
