# Groundwater Level and ET plots 

The following plots are produced to visualize a monitoring well's groundwater level and ET over time. It is meant to help the viewer look for patterns between the two.

### Load in Data

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
```

### Import Data

```python
# Create a groundwater file path
wilcox_fp = os.path.join("/capstone/aridgw/data/merged_timeseries_1km/et_gw_merged_willcox.csv")

# Read in groundwater data
wilcox_df = pd.read_csv(wilcox_fp)
wilcox_df.head()
```

```python

plt.figure(figsize=(12, 6))
plt.plot(wilcox_df['YEAR'], wilcox_df['WaterLevel_m'], 'o-', label='Water Level (m)')
# plt.plot(site_data['Year'], site_data['ET (Python)'], 's-', label='ET (Python)')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Depth to Groundwater (m)', fontsize=12)
plt.title('Depth to Groundwater Over Time In Wilcox, Arizona', fontsize=16)
plt.gca().invert_yaxis()  # Flip y axis
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(wilcox_df['YEAR'], wilcox_df['mean_ET_mm_yr'], 's-', label='Mean ET (mm/yr)', color='orange')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Mean ET (mm/yr)', fontsize=12)
plt.title('Mean Evapotranspiration Over Time In Wilcox, Arizona', fontsize=16)
plt.show()
```
