```python
import os
import earthaccess

earthaccess.login()  # will use your token automatically if set

```

```python
# Print the amount of files
results = earthaccess.search_data(
    short_name="MOD16A3GF",
    version="061",
)
print(len(results))

```

```python
earthaccess.download(
    results,
    local_path="data",
)
```
